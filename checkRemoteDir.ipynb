```python
import fsspec
from pathlib import Path

# 1. SET YOUR PATH HERE
# Use the same path you have in your NeMo YAML
uri = "hf://buckets/Tyl3rDrden/IvritKnessetRecordingsData"

print(f"Testing URI: {uri}")

try:
    # 2. Try to list files using fsspec (this is what Lhotse SHOULD do for remote)
    fs, fs_path = fsspec.core.url_to_fs(uri)
    all_remote_files = fs.ls(fs_path)
    
    print(f"\n--- FSSPEC Discovery ---")
    print(f"Found {len(all_remote_files)} items in bucket.")
    
    # Check for 'cuts' specifically
    remote_stems = [Path(p).stem.split('.')[0] for p in all_remote_files]
    if "cuts" in remote_stems:
        print("✅ SUCCESS: 'cuts' field found via fsspec!")
    else:
        print("❌ ERROR: 'cuts' NOT found in the bucket list.")
        print(f"Actual items found: {remote_stems[:5]}... (first 5)")

except Exception as e:
    print(f"❌ CONNECTION ERROR: {e}")

# 3. Verify why the code you shared failed
print(f"\n--- Local Pathlib Test ---")
local_test = list(Path(uri).glob("*"))
print(f"Path(uri).glob('*') found {len(local_test)} files.")
if len(local_test) == 0:
    print("💡 Note: Standard Pathlib cannot glob remote HF/S3 URLs. This is why _init_from_dir failed.")
```
